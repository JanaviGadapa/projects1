**Scraping: Web Scraping is an automatic way to retrieve unstructured data from a website and store them in a structured format.**

Notes: Scraping makes the website traffic spike and may cause the breakdown of the website server. Thus, not all websites allow people to scrape. How do you know which websites are allowed or not? You can look at the ‘robots.txt’ file of the website. You just simply put robots.txt after the URL that you want to scrape and you will see information on whether the website host allows you to scrape the website.

In python webscraping can be done through the library selenium. selenium is a web based driver that will automate website through a web browser of choice

**collecting multiple ads data in first page**

```python
url_elements=driver.find_elements(By.CLASS_NAME,'_3V_Ww')
```

```python
url_elements
```

```python
len(url_elements)
```

**Loop for Extracting urls one by one from url_elements**

```python
urls = []

for i in url_elements:
    urls.append(i.find_element(By.TAG_NAME,'a').get_property('href'))
```

```python
urls
```

**Defining empty columns**

```python
name=[]
price=[]
fuel=[]
driven_km=[]
transmision=[]
```

**Going through each and every url and finding data**

**Adding the data to empty columns**

```python
for i in urls:
    try:
        print(i)
        driver.get(i)
        print("Success")
        
        try:
            name.append(driver.find_element(By.CLASS_NAME,'_2iMMO').text)
        except:
            name.append("notfound")
        
        try:
            price.append(driver.find_element(By.CLASS_NAME, '_1uqlc').text)
        except:
            price.append("notfound")
        
        try:
            fuel.append(driver.find_elements(By.CLASS_NAME,"_3rMkw")[0].text)
        except:
            fuel.append("notfound")
        
        try:
            driven_km.append(driver.find_elements(By.CLASS_NAME,"_3rMkw")[1].text)
        except:
            driven_km.append("notfound")
        
        try:
            transmision.append(driver.find_elements(By.CLASS_NAME,"_3rMkw")[2].text)
        except:
            transmision.append("notfound")

    except:
        print(i)
        print("element not found")
        name.append("notfound")
        price.append("notfound")
        fuel.append("notfound")
        driven_km.append("notfound")
        transmision.append("not found")
                        
```

**Converting columns data into csv**

```python
import pandas as pd
data = pd.DataFrame({'Name':name,'Price':price,'Fuel_type':fuel,'Driven_km':driven_km,'Tranmission':transmision})
```

```python
data.to_csv("olx_cars1.csv")
```
